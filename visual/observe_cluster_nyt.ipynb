```python
# libraries
from pprint import pprint
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import json
# Data
json_file = json.load(open("../archives/gold_cluster.json", "r"))
pprint({k: len(v) for k, v in json_file.items()})
pprint({k: sum(pair[1] for pair in v) for k, v in json_file.items()})
pprint({k: sum(pair[1] for pair in v)/len(v) for k, v in json_file.items()})
raw_data = {k: [sum(pair[1] for pair in v)] for k, v in json_file.items()}
```

```python
raw_data
```

```python
r = [0]
df = pd.DataFrame(raw_data)
 
# From raw value to percentage
totals = [i+j+k+x+y for i,j,k,x,y in zip(df['HUMAN'], df['PLACE'], df['PARTY'], df['ORG'], df['OTHER'])]
HUMAN = [i / j * 100 for i,j in zip(df['HUMAN'], totals)]
PLACE = [i / j * 100 for i,j in zip(df['PLACE'], totals)]
PARTY = [i / j * 100 for i,j in zip(df['PARTY'], totals)]
ORG = [i / j * 100 for i,j in zip(df['ORG'], totals)]
OTHER = [i / j * 100 for i,j in zip(df['OTHER'], totals)]


```

```python
# plot
barWidth = 0.2
names = ('')
# Create green Bars
plt.bar(r, HUMAN,edgecolor='white', width=barWidth, label="HUMAN")
# Create orange Bars
plt.bar(r, PLACE, bottom=HUMAN, edgecolor='white', width=barWidth,  label="PLACE")
# Create blue Bars
plt.bar(r, PARTY, bottom=[i+j for i,j in zip(HUMAN, PLACE)],  edgecolor='white', width=barWidth, label="PARTY")
# Create blue Bars
plt.bar(r, ORG, bottom=[i+j+k for i,j,k in zip(HUMAN, PLACE, PARTY)], edgecolor='white', width=barWidth, label="ORG")
# Create blue Bars
plt.bar(r, OTHER, bottom=[i+j+k+x for i,j,k,x in zip(HUMAN, PLACE, PARTY, ORG)], edgecolor='white', width=barWidth, label="OTHER")
 
# Custom x axis
plt.xticks(r, names)
plt.xlabel("Data")
plt.ylabel("Percentage")
plt.title("Observe Cluster")
plt.legend()
 
# Show graphic
plt.show()
```
